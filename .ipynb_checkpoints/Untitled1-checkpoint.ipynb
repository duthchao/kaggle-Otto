```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn import ensemble
from sklearn import cross_validation
from sklearn import metrics
from sklearn import calibration
from time import time
import xgboost as xgb
from sklearn.grid_search import ParameterGrid
from datetime import datetime
```

```python
df_train = pd.read_csv('./data/train.csv')
df_test = pd.read_csv('./data/test.csv')
X = df_train.iloc[:,1:-1].values.astype(np.float64)
y = df_train.iloc[:,-1].map(lambda label:int(label.split('_')[1])-1).values.astype(np.float64)
df_all = pd.concat([df_train.iloc[:,1:-1],df_test.iloc[:,1:]],axis=0)
df_all.index = range(len(df_all))
```

```python
def make_sub(y_pred,result):
    print(result)
    result = '_'.join([str(item[0])+str(item[1]) for item in result.items()])
    df_sub = pd.DataFrame(index=range(1,y_pred.shape[0]+1),columns=['Class_'+str(i) for i in range(1,10)],data=y_pred)
    df_sub.index.rename('id',inplace=True)
    df_sub.to_csv(result+'.csv')
```

```python
def log_loss(y_prd,y):
    return metrics.log_loss(y_prd,y)
```

```python
_df = df_all
X = _df.iloc[:len(y)]
result = []
for i,(tr,va) in enumerate(cross_validation.KFold(len(y),shuffle=True,n_folds=5,random_state=1)):
    if i==1:
        break
    X_tr = X.iloc[tr]
    y_tr = y[tr]
    X_va = X.iloc[va]
    y_va = y[va]
    dtr = xgb.DMatrix(X_tr,y_tr,missing=np.NAN)
    dva = xgb.DMatrix(X_va,y_va,missing=np.NAN)
    param = {'objective':'multi:softprob','num_class':9,'eval_metric':'mlogloss',
             'eta':0.03,'max_depth':6,'gamma':0,'min_child_weight':4,'seed':1,
              'subsample':1.0,'colsample_bytree':1.0,'colsample_bylevel':1.}
    watchlist  = [(dtr,'tr'), (dva,'va')]
    num_round = 500
    bst = xgb.train(param, dtr, num_round, watchlist,early_stopping_rounds=10)
    tr_logloss = log_loss(y_tr,bst.predict(dtr))
    va_logloss = log_loss(y_va,bst.predict(dva))
    result.append({'n':i,'tr_loss':tr_logloss,'va_loss':va_logloss})
```

```python
result=[]
for p in ParameterGrid({'ss':[0.3,0.5,0.7],
                        'cs':[0.3,0.5,0.7]}):
    start = datetime.now()
    for i,(tr,va) in enumerate(cross_validation.StratifiedShuffleSplit(y,n_iter=1,train_size=0.7,test_size=0.3,random_state=1)):
        X_tr = X.iloc[tr]
        y_tr = y[tr]
        X_va = X.iloc[va]
        y_va = y[va]
        dtr = xgb.DMatrix(X_tr,y_tr,missing=np.NAN)
        dva = xgb.DMatrix(X_va,y_va,missing=np.NAN)
        param = {'objective':'multi:softprob','nthread':8,'silent':1,'num_class':9,'eval_metric':'mlogloss',
                 'eta':0.1,'max_depth':6,'gamma':0,'min_child_weight':1,'seed':1,
                  'subsample':p['ss'],'colsample_bytree':p['cs'],'colsample_bylevel':1.}
        watchlist  = [(dtr,'tr'), (dva,'va')]
        num_round = 1000
        bst = xgb.train(param, dtr, num_round, watchlist,early_stopping_rounds=10)
        tr_logloss = log_loss(y_tr,bst.predict(dtr,ntree_limit=bst.best_iteration))
        va_logloss = log_loss(y_va,bst.predict(dva,ntree_limit=bst.best_iteration))
        p['tr'] = tr_logloss
        p['va'] = va_logloss
        p['t'] = str(datetime.now() - start)
        print p
        result.append(p)
```
